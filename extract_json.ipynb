```python
from llama_cloud_services import LlamaExtract
from pydantic import BaseModel, Field

# Define schema using Pydantic
class Resume(BaseModel):
    name: str = Field(description="title of the page")
    git_id: str = Field(description="would have the word 'github.com'")
    linkedin_id: str = Field(description="would have the word 'linkedin.com'")
    email_id: str = Field(description="would have the word 'gmail.com'")
    skills: list[str] = Field(description="list of content inside the main heading 'Skills'")
    Projects: list[str] = Field(description="list of content inside the main heading 'Projects' extracted from numbered, roman numeral, or bulleted format")
    Acheivements: list[str] = Field(description="list of content inside the main heading Acheivement's extracted from numbered, roman numeral, or bulleted format")
    education: str = Field(description="list of content inside the main heading 'Education'")
    description: str = Field(description="the longest paragraph in the document")
    phone_number: str = Field(description="starting with +91")
    work_experience: list[str] = Field(description="list the work experience")

# Create extraction agent
agent = extractor.create_agent(name="resume_final", data_schema=Resume)   #here resume refers to the class defined above
```
